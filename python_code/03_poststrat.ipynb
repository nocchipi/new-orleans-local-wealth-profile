```python
import os
import polars as pl
```

```python
# create output directory
OUTPUT_PATH = os.getenv("OUTPUT_PATH")
os.makedirs(OUTPUT_PATH,exist_ok=True)
OUTPUT_PATH
```

```python
STATEFIP_CODE = int(os.environ.get("STATEFIP_CODE"))
PUMA_CODES = [int(item.strip()) for item in os.environ.get("PUMA_CODES").split(',') if item]
print(f"STATEFIP_CODE: {STATEFIP_CODE}\nPUMA_CODES: {PUMA_CODES}")
```

```python
ipums_post_strat = pl.read_parquet(os.path.join(OUTPUT_PATH,f"ipums_post_stratification_{STATEFIP_CODE}.parquet"))
ipums_post_strat
```

```python
ipums_post_strat.describe()
```

```python
ipums_post_strat = ipums_post_strat.with_columns(
    pl.col("PUMA").cast(pl.Utf8),
    pl.col("COUNTYFIP").cast(pl.Utf8),
    pl.col("state").cast(pl.Utf8),
)
ipums_post_strat
```

```python
ipums_expanded = (ipums_post_strat
 .with_row_index("id",offset=1)
 .with_columns(
    pl.all().repeat_by(pl.col("WGT"))
 )
 .explode(pl.all())
)
ipums_expanded 
```
